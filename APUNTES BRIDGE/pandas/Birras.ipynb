## Ejercicio 23
Consigue en un DataFrame todas las cervezas en cuyas cervecerías el mínimo nivel de `abv` sea de 0.07.

```python
df_min_abv = df.groupby('brewery_id')['abv'].min()

df_filtered = df_min_abv[df_min_abv >= 0.07]

df_result = df[df['brewery_id'].isin(df_filtered.index)]
df_result
```

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media

```python
global_mean = new_dataset['ibu'].mean()

state_means = new_dataset.groupby('state')['ibu'].mean()

states_above_mean = state_means[state_means > global_mean].index.tolist()

beers_above_mean = new_dataset[new_dataset['state'].isin(states_above_mean)]

beers_above_mean
```
